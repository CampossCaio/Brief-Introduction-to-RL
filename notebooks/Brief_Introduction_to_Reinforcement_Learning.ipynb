## Theory



### What is Reinforcement Learning?

Reinforcement Learning is a type of machine learning where an agent learns to make decisions by performing actions in an environment to maximize cumulative reward. Unlike supervised learning, where the model is trained on a dataset with labeled inputs and outputs, RL involves learning through trial and error, with the agent receiving feedback in the form of rewards or penalties.


  ![RL loop](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/rl_loop.png)


### Key Components of Reinforcement Learning



- **Agent**: The learner or decision-maker.
- **Environment**: The external system with which the agent interacts.
- **State**: A representation of the current situation in the environment.
- **Action**: A decision made by the agent that affects the environment.
- **Reward**: Feedback from the environment used to evaluate the action's effectiveness.

### How Reinforcement Learning Works

In Reinforcement Learning (RL), the agent performs an action based on its current state and receives feedback from the environment in the form of a reward. The objective is to learn a policy that maximizes the cumulative reward over time. For instance, let's consider our RL agent learning to play Sonic. The learning process can be modeled as an iterative loop that functions as follows:

![Rl loop Sonic Example](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/2.png)

- The RL agent begins by receiving an initial state $S_0$ from the environment, such as a snapshot from the game Sonic.
- Based on this state $S_0$, the agent takes an action $A_0$, for instance, moving right. Initially, this action is chosen randomly.
- Following this action, the environment updates to a new state $S_1$ (a subsequent frame from the game).
- The environment then provides a reward $R_1$ to the agent, which might be a $+1$ if the agent is still alive.

This RL loop outputs a sequence of **state**, **action**, **reward** and **next state**. $S_0, A_0, R_1, S_1$

### Reward Maximization


The RL agent operates on the principle of maximizing rewards. This means that the agent aims to take the best possible actions to achieve the highest cumulative reward.

The total rewards accumulated over time for each action taken can be expressed as:

$$
G_t = \sum_{k=0}^{T} \gamma^k R_{t+k+1}
$$

or

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots + \gamma^{k} R_{t+k+1}
$$




Imagine our RL agent (a squirrel) navigating an area filled with nuts and traps. The objective is to gather as many nuts as possible while avoiding getting caught by traps.

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/1.png)

### Discounting rewards

Discounting rewards in reinforcement learning involves assigning less value to future rewards compared to immediate ones. This is typically represented using a discount factor, denoted by ${\gamma}$, where ${0 < \gamma < 1}$.




In the context of our squirrel navigating an environment with nuts and traps, discounting rewards would mean that the squirrel places higher importance on immediate rewards (such as finding and eating a nearby nut) compared to rewards that might be obtained in the distant future.


So, our total expected rewards, taking into account discounting, are:

$$
G_t = \sum_{k=0}^{T} \gamma^k R_{t+k+1} where\quad\gamma \in [0, 1]
$$

### Markov Property


The Markov Property is fundamental in Markov Decision Processes (MDPs), a mathematical framework used in reinforcement learning and sequential decision-making under uncertainty. **This property describes the dependency between states in a sequence of events, where the future state depends only on the current state, not on past states.** In other words, given the current state, the future state is conditionally independent of past states.

Formally, it can be expressed as:

$$ P(s_{t+1} \mid s_t, s_{t-1}, \ldots, s_0) = P(s_{t+1} \mid s_t) $$

where $S_t$ represents the state at time $t$ and $P$ denotes conditional probability. This formulation implies that once the current state is known, it is not necessary to remember all past states to predict the next state.

### Tasks and their types in reinforcement learning

In reinforcement learning (RL), a "task" refers to the specific goal or objective that an agent aims to achieve through interaction with an environment. Tasks in RL can vary widely based on the problem domain and the desired outcome.

#### Episodic Tasks:

- In episodic tasks, the agent interacts with the environment for a finite number of steps or episodes.
- Each episode has a clear beginning and end, and the agent's goal is often to maximize the total reward accumulated within each episode.

   ![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/episodic_tasks.png)



#### Continuous Tasks:
- Continuous tasks involve ongoing interaction with the environment without predefined episodes.
- The agent operates indefinitely, and the goal typically involves maximizing the long-term cumulative reward over an extended period.

  ![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/continuous_task.jpeg)

### Exploration and Explotation



Exploration in reinforcement learning involves seeking new information to understand the environment better, often through random or uncertain actions. Exploitation focuses on using existing knowledge to maximize immediate rewards by making decisions based on learned strategies or policies. Balancing these strategies is crucial: too much exploration can slow learning, while excessive exploitation may miss better long-term strategies. Finding the right balance allows RL agents to effectively learn and make optimal decisions in dynamic environments.

This helps agents like our squirrel understand where rewards (nuts) are located and where risks (traps) may be present. Exploitation, on the other hand, uses existing knowledge to maximize immediate rewards based on learned strategies or policies. Balancing these strategies is crucial: effective exploration gathers essential data, while exploitation optimizes decisions to achieve the highest long-term rewards.

### Approaches to Reinforcement Learning


There are several approaches to reinforcement learning (RL), each employing unique techniques and algorithms to specific types of problems and environments. Here, we will focus on the two main approaches: <br><br>

####Value-Based Methods
In value-based reinforcement learning, the agent aims to maximize the value function $V(s)$, which predicts the maximum expected future reward the agent can achieve from each state $S$. This function serves as a guide for the agent to make decisions that lead to higher cumulative rewards over time.


The value of a state represents the total expected reward that an RL agent anticipates collecting over the future, starting from that specific state.

\begin{equation}
V(s) = \mathbb{E} \left[ \sum_{t=0}^{\infty} \gamma^t R_{t+1} \mid s_0 = s \right]
\end{equation}

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/4.png)

<br>
<br>

#### Policy-Based Methods
In policy-based reinforcement learning, we focus on optimizing a policy, which dictates the agent's behavior in the environment. The policy specifies the actions the agent should take based on the observed states to maximize its cumulative reward over time. Unlike value-based methods that estimate the value function directly, policy-based approaches directly learn and improve the policy through iterative updates to enhance decision-making effectiveness in various environments.
<br><br>

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/3.png)

Delving deeper into policies, we categorize them into two distinct types:

- Deterministic Policy: A deterministic policy, when given a particular state $s$, always produces the same action $a$. In other words, it maps states directly to actions without variability: $\pi(s)=a$.

- Stochastic Policy: Conversely, a stochastic policy provides a probability distribution over possible actions given a state $s$. This means it defines the likelihood of taking each action: $\pi(a | s) = P(A = a | S = s)$.




### Let's recap

- Reinforcement Learning (RL) is a computational approach where an agent learns by interacting with its environment, receiving rewards as feedback—either positive or negative.

- The main goal of any RL agent is to maximize its expected cumulative reward, following the reward hypothesis that frames all objectives as maximizing this cumulative reward.

- The RL process operates in a loop, generating sequences of states, actions, rewards, and subsequent states.

- To calculate the expected cumulative reward (or return), rewards are discounted, prioritizing nearer-term rewards due to their higher predictability compared to distant future rewards.

- In solving an RL problem, the aim is to determine an optimal policy. This policy guides the agent's actions based on its current state, maximizing the expected return.

- Two main approaches to finding the optimal policy are:

   - Policy-based methods: Directly train the policy to maximize expected return.
   - Value-based methods: Train a value function to estimate expected return for each state, then derive the optimal policy based on these values.

```python
import imageio
from IPython.display import HTML
from base64 import b64encode
import time


def generate_video_from_frames(out_dir, frames, fps=4):
  imageio.mimsave(out_dir, [np.array(img) for i, img in enumerate(frames)], fps=fps)

def display_video(file_path):
  video = open(file_path, 'rb').read()
  data_url = "data:video/mp4;base64," + b64encode(video).decode()

  return HTML(f'<video width="400" height="400" controls><source src="{data_url}" type="video/mp4"></video>')

def generate_and_display_video(frames, out_dir):
   generate_video_from_frames(out_dir, frames)
   return display_video(out_dir)


def plot_frame(env):
  frame = env.render()
  plt.axis('off')
  plt.imshow(frame);


def test_agent(env, policy, n_test_episodes=10):
    frames = []
    video_path = 'replay.mp4'

    for episode in range(n_test_episodes):
        state, info = env.reset()
        terminated = False
        truncated = False
        frames.append(env.render())

        while not terminated or truncated:
            p = policy(state)
            if isinstance(p, np.ndarray):
                action = np.random.choice(4, p=p)
            else:
                action = p
            next_state, reward, terminated, truncated, info = env.step(action)
            img = env.render()
            frames.append(img)
            state = next_state

    return generate_and_display_video(frames, video_path)

def evaluate_agent(env, n_eval_episodes = 100, max_steps = 99):
  episode_rewards = []
  for episode in tqdm(range(n_eval_episodes)):

    state, info = env.reset()
    step = 0
    truncated = False
    terminated = False
    total_rewards_ep = 0

    for step in range(max_steps):
      # Take the action (index) that have the maximum expected future reward given that state
      action = greedy_policy(state)
      new_state, reward, terminated, truncated, _ = env.step(action)
      total_rewards_ep += reward

      if terminated or truncated:
        break
      state = new_state
    episode_rewards.append(total_rewards_ep)
  mean_reward = np.mean(episode_rewards)
  std_reward = np.std(episode_rewards)

  return mean_reward, std_reward, episode_rewards
```

#### Create the $Q(s, a)$ table

A Q-Table is essentially a lookup table where we calculate the maximum expected future rewards for each action at every state. This table serves as a guide to help us determine the best action to take in any given state.

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/6.png)

```python
q_table = np.zeros((env.observation_space.n, env.action_space.n))
q_table
```

#### Create the greedy policy

A greedy policy involves always selecting the action with the highest estimated reward for a given state, prioritizing exploitation of known good actions.

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/greedy_policy.png)

```python
def greedy_policy(state):

  return np.argmax(q_table[state])
```

#### Create epsilon-greedy policy

The epsilon greedy policy establishes a balance between exploration and exploitation: with probability 1-ε, it explores the best-known action, while with probability ε, it explores a random action to potentially discover better strategies.

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/8ffea4a57a513dc385a1f58a0450f120e3241b1d/images/e_greedy_policy.png)

```python
def epsilon_greedy_policy(state, epsilon = 1):
  random_num = random.uniform(0, 1)

  if random_num > epsilon:
    return greedy_policy(state) # Exploit
  else:
    return env.action_space.sample() # Explore
```

### Implement the algorithm

![](https://raw.githubusercontent.com/CampossCaio/Brief-Introduction-to-RL/main/images/q_learning_steps.png)

```python
def q_learning(env, q_table, n_episodes, alpha = 0.1, gamma = 0.99, epsilon = 1.0, epsilon_decay = 0.999, min_epsilon = 0.01):

  for episode in tqdm(range(n_episodes)):
    state, info = env.reset()
    terminated = False
    truncated = False

    while not terminated or truncated:
      action = epsilon_greedy_policy(state, epsilon)

      next_state, reward, terminated, truncated, info = env.step(action)
      best_next_action = greedy_policy(next_state)

      qsa = q_table[state][action]
      next_qsa = q_table[next_state][best_next_action]

      q_table[state][action] = qsa + alpha * (reward + gamma * (next_qsa - qsa))
      state = next_state

    epsilon = max(min_epsilon, epsilon * epsilon_decay)
```

#### Define the hyperparameters

```python
# Environment
gamma = 0.99  # Discount factor
max_steps = 99 # Max steps per episode

# Training
alpha = 0.1  # Learning rate
n_train_pisodes = 10000 # Total training episodes

# Evaluating
n_eval_episodes = 100 # Total evaluation episodes

# Exploration/ Explotation
epsilon = 1.0  # Exploration rate at start
epsilon_decay = 0.99 # Decay rate for exploration prob
min_epsilon = 0.01 # Minimum exploration probability
```

#### Train

```python
q_learning(env, q_table, n_episodes=n_train_pisodes)
```
